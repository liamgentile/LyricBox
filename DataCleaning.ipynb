# Folk Lyrics

I'll start by reading in the csv.

```python
folk_df = pd.read_csv('/Users/liamgentile/Desktop/lyric_files/folk_lyrics.csv', converters={'lyrics': eval})
```

```python
# display the first 5 rows
folk_df.head(5)
```

We have an additional, unnecessary index, so I'll drop that using df.drop()

```python
folk_df.drop('Unnamed: 0', axis=1, inplace=True)
folk_df
```

One possible thing I should check for is cover songs. I'll look for this by looking up song names that contain the word cover, as AZ-lyrics combines "cover" into the song name.

```python
folk_df[folk_df['song'].str.contains('cover')]
```

It seems like the only real cover here is someone you loved cover by passenger. So i'll remove that.

```python
folk_df = folk_df[~folk_df['song'].str.contains('lewiscapaldicover')]
folk_df
```

Now I'll check for remixes with the same str.contains method.

```python
folk_df[folk_df['song'].str.contains('remix')]
```

These all seem to be remixes, so I'll go ahead and remove them now by assigning the dataframe to everything except these songs (using ~). 

```python
folk_df = folk_df[~folk_df['song'].str.contains('remix')]
folk_df
```

Another cleaning step is dropping duplicates. To do this I can simply use pandas' drop_duplicates.

```python
folk_df.drop_duplicates(subset="song", keep=False, inplace=True)
```

```python
folk_df
```

In order to look within the lyrics, I have to add another column which is the lyrics column converted to a string. That way I can actually look up the content within the lyrics.

```python
folk_df['lyrics_string'] = folk_df['lyrics'].apply(lambda x: ','.join(map(str, x)))
```

```python
folk_df
```

I noticed that songs that contain 'Lyrics' within the lyrics themselves have serious data quality issues, with a lot of additional text and the beginning and end of the songs. So, I'll drop these, as I can't automatically correct them. 

```python
folk_df[folk_df['lyrics_string'].str.contains('Lyrics')]
```

```python
folk_df = folk_df[~folk_df['lyrics_string'].str.contains('Lyrics')]
```

```python
folk_df
```

Now we have 6726 songs left. I'll now look up songs that have square brackets in the lyrics. This is usually added to indicate song section or sometimes who is singing.

```python
folk_df[folk_df['lyrics_string'].str.contains(']')]
```

605 songs have this issue! Rather than removing them I can take advantage of the fact that I have the lyrics separated into lists of lines. Using the following for loop I can remove any array (comma separated value) which contains a square bracket. 

```python
for lists in folk_df['lyrics']:
    for x in lists:
        if ']' in x:
            lists.remove(x)
```

Let's check one of the examples. Number 299 had [Spoken] at the beginning. Let's see if it fixed it.

```python
folk_df['lyrics'][299]
```

Looks good!

Now I am going to remove lines that contain potentially offensive language (i.e. common swear words, words associated with sexism and racism). I have replaced such language with X in the following code.
